```python
df = df[df['Row ID'].notna()]
```

```python
df['Row ID'] = df['Row ID'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df['Postal Code'] = df['Postal Code'].astype(int)
df['Quantity'] = df['Quantity'].astype(int)
```

```python
df.dtypes
```

```python
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
df[["Order Date", "Ship Date", "Shipping Days"]].head()
```

```python
df["Profit Margin (%)"] = (df["Profit"] / df["Sales"]) * 100
```

```python
df[["Sales", "Profit", "Profit Margin (%)"]].head()
```

```python
def sales_category(sales):
    if sales <= 100:
        return "Low"
    elif sales <= 500:
        return "Medium"
    else:
        return "High"

df["Sales Category"] = df["Sales"].apply(sales_category)
```

```python
df[["Sales", "Sales Category"]].head(10)
```

```python
df["Profit Status"] = df["Profit"].apply(lambda x: "Profit" if x >= 0 else "Loss")
```

```python
df[["Profit", "Profit Status"]].head(10)
```

```python
def get_quarter(month):
    if month in [1, 2, 3]:
        return "Q1"
    elif month in [4, 5, 6]:
        return "Q2"
    elif month in [7, 8, 9]:
        return "Q3"
    else:
        return "Q4"

df["Order Quarter"] = df["Month"].apply(get_quarter)
```

```python
region_sales = (
    df.groupby("Region")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

region_sales
```

## Business Insight #1:
### The West region generated the highest sales (725,457.82), while the South region recorded the lowest sales (391,721.91). This indicates that the West region is the strongest revenue contributor and the South region may require targeted marketing or sales improvement strategies.

```python
region_profit = (
    df.groupby("Region")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

region_profit
```

## Business Insight 2

## • West region generated both the highest sales (₹725,457.82) and the highest profit (₹108,418.45), making it the strongest performing region.

## • Central region generated higher sales than South but earned the lowest profit, indicating potential issues such as excessive discounts or low-margin products.

## • South generated lower sales but higher profit than Central, suggesting better profitability.

```python
category_profit = (
    df.groupby("Category")[["Sales", "Profit"]]
      .sum()
      .sort_values(by="Profit", ascending=False)
)

category_profit
```

## Business Insight 3

### Technology generated the highest sales (₹836,154.03) and the highest profit (₹145,454.95), making it the company's strongest-performing category.

### Office Supplies produced lower sales than Furniture but significantly higher profit, indicating better profitability.

### Furniture generated high sales but very low profit (₹18,451.27), suggesting potential issues such as high discounts, shipping costs, or low-margin products.

```python
subcategory_profit = (
    df.groupby("Sub-Category")[["Sales", "Profit"]]
      .sum()
      .sort_values(by="Profit")
)

subcategory_profit
```

## Business Insight 4

### Tables generated ₹206,965.53 in sales but resulted in a loss of ₹17,725.48, making it the least profitable sub-category.

### Bookcases and Supplies also generated negative profits.

### Copiers, Phones, and Accessories were the most profitable sub-categories and should be prioritized in future sales and marketing strategies.

```python
top_products = (
    df.groupby("Product Name")["Profit"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products
```

## Business Insight #5

### The Canon imageCLASS 2200 Advanced Copier generated the highest profit (₹25,199.93), significantly outperforming all other products.

### Most of the top 10 profitable products belong to the Technology category, confirming that Technology is the company's strongest-performing business segment.

```python
bottom_products = (
    df.groupby("Product Name")["Profit"]
      .sum()
      .sort_values()
      .head(10)
)

bottom_products
```

## Business Insight #6

### The Cubify CubeX 3D Printer Double Head Print recorded the highest loss (-₹8,879.97), making it the least profitable product.

### Several table-related products also appeared among the top loss-making products, confirming that the Tables sub-category is negatively impacting overall profitability.

### The business should review pricing, discount policies, and procurement costs for these products.

```python
monthly_sales = (
    df.groupby("Month")["Sales"]
      .sum()
      .sort_index()
)

monthly_sales
```

```python
df["Month"].value_counts().sort_index()
```

```python
# Create month number from Order Date
df["Order Month"] = df["Order Date"].dt.month

# Create month name
df["Month Name"] = df["Order Date"].dt.month_name()

# Create year (from Order Date)
df["Order Year"] = df["Order Date"].dt.year
```

```python
df[["Order Date", "Order Month", "Month Name", "Order Year"]].head()
```

```python
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

monthly_sales = (
    df.groupby("Month Name")["Sales"]
      .sum()
      .reindex(month_order)
)

monthly_sales
```

## Business Insight #7

### Sales showed a clear seasonal trend, with November (₹352,461.07) and December (₹325,293.50) being the strongest-performing months.

### February recorded the lowest sales (₹59,751.25).

### The company should prepare inventory and marketing campaigns before the high-demand period from September to December.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker='o',
    linewidth=3
)

# Display sales value on each point
for i, value in enumerate(monthly_sales.values):
    plt.text(i, value + 5000, f'{value/1000:.0f}K', ha='center', fontsize=9)

plt.title("Monthly Sales Trend", fontsize=18, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales", fontsize=12)

plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
segment_analysis = (
    df.groupby("Segment")[["Sales", "Profit"]]
      .sum()
      .sort_values(by="Sales", ascending=False)
)

segment_analysis
```

## Business Insight #8

### The Consumer segment generated the highest sales (₹1.16M) and profit (₹134K), making it the company's most valuable customer segment.

### Corporate customers ranked second in both sales and profit, while the Home Office segment contributed the least.

### The company should continue investing in Consumer and Corporate segments while exploring strategies to grow Home Office sales.

```python
discount_analysis = (
    df.groupby("Discount")[["Sales", "Profit"]]
      .mean()
      .sort_index()
)

discount_analysis
```

## Business Insight #9

### The analysis revealed a strong negative relationship between discount percentage and profitability.

### Products sold with discounts above 30% consistently generated negative average profits.

### The company should optimize its discount strategy by limiting excessive discounts and focusing on targeted promotional campaigns.

```python
df[["Sales", "Profit", "Discount", "Quantity", "Shipping Days"]].corr()
```

## Business Insight #10

### Correlation analysis showed that:

### • Sales and Profit have a moderate positive correlation (0.48).

### • Discount has a negative correlation (-0.22) with Profit, confirming that higher discounts reduce profitability.

### • Quantity has only a weak positive relationship with Sales.

### • Shipping Days has almost no correlation with Sales or Profit.
